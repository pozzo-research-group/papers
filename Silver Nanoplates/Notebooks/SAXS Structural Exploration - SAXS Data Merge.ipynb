```python
import sys
sys.path.append('../Data_Merge/')
import data_merge
sys.path.append('../sasview/src')
import os 
import numpy as np
from sasmodels.data import load_data
import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings("ignore")
```

## Open File Directory

```python
folders = os.listdir('../SAXS Structural Exploration/SAXS/Raw Data')
folders
```

## Merge Data 

```python
for i in range(136):
    direc = '../SAXS Structural Exploration/SAXS/Raw Data' + str(folders[i - 1]) + '/'
    filenames = os.listdir(direc)
    ESAXS = filenames[0]
    SAXS = filenames[1]
    MAXS = filenames[2]
    cut_locations = [20, 200] # This array specifies where the merge occurs 
    ESAXS = load_data(direc + ESAXS)
    SAXS = load_data(direc + SAXS)
    MAXS = load_data(direc + MAXS)
    idir = ''
    data = data_merge.perform_merge(Data_1 = ESAXS, Data_2 = SAXS, Data_3 = MAXS, cut_locations = cut_locations, idir = idir, save = False)
    lower = 3 # This specifies the lower limit of where to cut the data
    upper = -(data.shape[0] - 250) #This specifies the upper limit of where to cut the data
    plt.loglog(data[lower:upper,0], data[lower:upper,1])
    plt.ylabel('Intensity (a.u.)')
    plt.xlabel('q ($\AA^{-1}$)')
    plt.title('i = ' + str(i) + '| Merged Data: ' + filenames[0])
    #np.savetxt('../SAXS Structural Exploration/SAXS/Merged Data' + str(folders[i - 1]) + '.dat', data[lower:upper,:]) #save merged data as txt
    #np.save('../SAXS Structural Exploration/SAXS/Merged Data' + str(folders[i - 1]), data[lower:upper,:]) #save merged data as npy 
```
