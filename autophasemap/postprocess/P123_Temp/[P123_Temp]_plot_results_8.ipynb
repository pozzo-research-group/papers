```python
def setup_axs():
    fig, axs = plt.subplots(2,4, figsize=(4*4, 4*2))
    fig.subplots_adjust(wspace=0.5, hspace=0.5)
    axs = axs.flatten()
    for k in range(8):
        axs[k].text(0.1, 0.9, r'(%s)'%string.ascii_uppercase[k], 
        transform=axs[k].transAxes, 
        size=20, weight='bold'
        )
    
    return fig, axs
```
